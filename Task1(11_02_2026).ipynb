```python
# 1 User Login Check
# Stored username and password
username = "admin"
password = "1234"

# User input
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Login check
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")

```

```python
# 2 Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

```python
# 3 Simple Data Cleaner
# Given list
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean each name
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print cleaned list
print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
# 4 Message Length Analyzer
# Given messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Status: Long Message")
    else:
        print("Status: Short Message")

    print("-------------------")
```

```python
# 5 Error Message Detector
# Given logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Check each log entry
for log in logs:
    if log == "ERROR":
        error_count += 1

# Print final result
print("Total Log Entries:", len(logs))
print("Total ERROR Entries:", error_count)


```
